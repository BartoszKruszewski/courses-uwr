```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import plotly.graph_objects as go
```

```python
np.random.seed(0)
X = np.random.uniform(-10, 10, 100)
epsilon = np.random.normal(0, 0.1, 100)
y = 3 * X + 4 + epsilon
```

```python
X_tensor = torch.tensor(X, dtype=torch.float32).view(-1, 1)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1) 
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)
    
model = LinearRegressionModel()
```

```python
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
num_epochs = 1000
for epoch in range(num_epochs):
    y_pred = model(X_tensor)
    loss = criterion(y_pred, y_tensor)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
predicted = model(X_tensor).detach().numpy()

scatter = go.Scatter(x=X, y=y, mode='markers', name='True data')
line = go.Scatter(x=X, y=predicted.flatten(), mode='lines', name='Fitted line', line=dict(color='red'))

layout = go.Layout(
    title="Linear Regression Fit",
    xaxis=dict(title="X"),
    yaxis=dict(title="y"),
    showlegend=True
)

fig = go.Figure(data=[scatter, line], layout=layout)
fig.show(config=dict(scrollZoom=True))

print(f"Learned parameters: W = {model.linear.weight.item():.4f}, b = {model.linear.bias.item():.4f}")
```
